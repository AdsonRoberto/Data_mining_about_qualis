##### Gerando gráfico do número de publicações

```python
df = df2.groupby(['Area'])['Títulos '].sum()
df = df.to_frame()
```

```python
areas_titulos = df.values.tolist()
areas_titulos
```

```python
# Make a list of the people, we'll use it later
cursos = list(df2['Área de avaliação'].unique())
```

```python
dict(zip(cursos, cursos))
```

```python
[area for area in areas]
```

```python
import networkx as nx
g = nx.from_pandas_edgelist(df2, source='Área de avaliação', target='Area')
```

```python
[g.degree(area) for area in areas]
```

```python
club_size = [g.degree(area) * num_titulos for area, num_titulos in areas_titulos]
```

```python
club_size
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

# 1. Create the graph
g = nx.from_pandas_edgelist(df2, source='Área de avaliação', target='Area') 

# 2. Create a layout for our nodes 
layout = nx.spring_layout(g, k=0.2,iterations=50)

# 3. Draw the parts we want
# Edges thin and grey
# People small and grey
# Clubs sized according to their number of connections
# Clubs blue
# Labels for clubs ONLY
# People who are highly connected are a highlighted color

# Go through every club name, ask the graph how many
# connections it has. Multiply that by 80 to get the circle size
club_size = [num_titulos for area, num_titulos in areas_titulos]
nx.draw_networkx_nodes(g, 
                       layout, 
                       nodelist=areas, 
                       node_size=club_size, # a LIST of sizes, based on g.degree
                       node_color='lightblue')

# Draw EVERYONE
nx.draw_networkx_nodes(g, layout, nodelist=cursos, node_color='orange', node_size=100)
node_labels = dict(zip(cursos, cursos))
nx.draw_networkx_labels(g, layout, font_size=7, labels=node_labels)

# Draw POPULAR Course
popular_curso = [curso for curso in cursos if g.degree(curso) > 1]
nx.draw_networkx_nodes(g, layout, nodelist=popular_curso, node_color='orange', node_size=100)

nx.draw_networkx_edges(g, layout, width=1, edge_color="orange")

node_labels = dict(zip(areas, areas))
nx.draw_networkx_labels(g, layout, labels=node_labels)

# 4. Turn off the axis because I know you don't want it
plt.axis('off')

plt.title("Análise da Quantidade de Títulos por Área")

# 5. Tell matplotlib to show it
plt.show()
```
